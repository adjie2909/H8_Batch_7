# Nested While Loops

```python
age = 29
gender = 'M'

if age < 18:
    if gender == 'M':
        print('son')
    else:
        print('daughter')
elif age >= 18 and age < 65:
    if gender == 'M':
        print('Father')
    else:
        print('Mother')
else:
    if gender == 'M':
        print('grandfather')
    else:
        print('grandmother')
```

# Contoh Penggunaan

```python
temp         = input('ketikkan Temperatur yang ingin dikonversim, eg. 45F, 120C:')
degree       = int(temp[:-1])
i_convertion = temp[-1]

if i_convertion   == "C":
    result   = int(round((9 * degree) / 5 + 32))
elif i_convertion == "F":
    result   = int(round((degree - 32) * 5 / 9))
else:
    print("Masukkan input yang benar")
print(f"Temperaturnya adalah {result} derajat")
```
